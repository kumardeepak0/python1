#### Take

- take only specific index values
- inside take ,one more argument is axis
- the default value of axis :0
- axis =0 means rows
- which means 100,200,300, rows will come
- axis = 1 means column
- which means 9, 10 , 12 columns will come 

```python
visa_df.take([100,200,300])
```

```python
visa_df.take([100,200,300],axis=1)
```

```python
visa_df.take([3,5,7],axis=1)
```

```python
list1=[2,3,7]
visa_df.take(list1,axis=1)
```

```python
visa_df.take([2,5,8],axis=1)

# in python index start with =0
# 2===== 3rd column
# 5====== 6th column
# 8======= 9th column
```

```python
visa_df.take([2,5,8],axis=0)
```

```python
visa_df.take((2,5,7),axis=1)
```
