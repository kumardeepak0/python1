- How to drop any Rows and Column

```python
# what is your data frame name : name
************* axis*****************
# I want to drop column
# axis=1 ============= column
# axis=0 ============= rows
******************************************************
# what is the name of the column means
*********** inplace*************************
#   you are modifying the table
#  the dataframes overwiter or
# you want to create a dataframes


# inplace=True  : data will be overwrite
```

```python
name.drop('city',axis=1,inplace=True)
name
```

```python
name.drop('D',axis=0,inplace=True)
name
```
